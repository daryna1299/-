## 1 Логнормальний розподіл

Більш детальна інформація в [документації](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.lognorm.html).

```python
from scipy.stats import lognorm
```

Випадкова велиина $Y$ має логнормальний розподіл, якщо $Y = \exp \left\{ X \right\}$, де $X$ - випадкова величина з нормального розподілу.

```python
mu = 0
sigma = 1
```

Параметри $s, loc, scale$ відповідають логнормальному розподілу зі щільністю $$f(x) = \frac{1}{s (x - loc) \sqrt{2 \pi}} \exp \left\{ -\frac{1}{2 s^2} \log^2 \left( \frac{x - loc}{scale} \right) \right\}.$$

```python
s = sigma
loc = 0
scale = np.exp(mu)
```

### 1.1 Щільність

```python
x1 = np.arange(-1, 5, 0.1)
```

Виведіть значення функції щільності логнормального розподілу з параметрами $s, loc, scale$ в точках вектора $x1$.

```python
# Місце для Вашого коду
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm
sigma = 1
mu = 0
s = sigma
loc = 0
scale = np.exp(mu)
x1 = np.arange(-1, 5, 0.1)
y1 = lognorm.pdf(x1, s, loc=loc, scale=scale)
print(y1)
```

```python
# Місце для Вашого коду
x2 = np.arange(-1, 5.01, 0.01)
y2 = lognorm.pdf(x2, s, loc=loc, scale=scale)
plt.plot(x2, y2)
plt.show()
```

### 1.2 Функція розподілу

Виведіть значення функції розподілу логнормального розподілу з параметрами $s, loc, scale$ в точках вектора $x1$.

Зобразіть графічно значення функції розподілу логнормального розподілу з параметрами $s, loc, scale$ в точках вектора $x2$.

```python
loc = 0
scale = 1
```

### 3.1 Рівномірний розподіл

Зобразіть графічно функцію щільності, функцію розподілу та нормовану гістограму $100000$ випадкових величин з параметром `bins=50` разом з функцією щільності рівномірного розподілу з параметрами:
1. $a = 0, b = 2$
2. $a = 0, b = 5$
3. $a = 0, b = 10$
4. $a = -2, b = 2$
5. $a = -5, b = 5$

Зробіть шкали на відповідних графіках однаковими для кращого порівняння. Наприклад, усі графіки щільності повинні мати однакові границі як по осі $x$, так і по осі $y$.

```python
# Місце для Вашого коду
from scipy.stats import uniform
def rivn (a, b):
    loc = a
    scale = b - loc
    fig, axs = plt.subplots(1,3, figsize=(20,5))
    print(fig)
    
    x2 = np.arange(a-1, b+1, 0.01)
    y2 = uniform.pdf(x2, loc=loc, scale=scale)

    axs[0].plot(x2, y2, 'r')
    axs[0].set_title('pdf')
    axs[0].set_ylim(a, b)
    axs[0].set_xlim(a, b)
    y21 = uniform.cdf(x2, loc=loc, scale=scale)

    axs[1].plot(x2, y21, 'g-')
    axs[1].set_title('cdf')
    axs[1].set_ylim(a, b)
    axs[1].set_xlim(a, b)
    unif_rvs = uniform.rvs(loc=loc, scale=scale, size=100000)
    
    
    axs[2].plot(x2, y2, 'r-.')
    axs[2].set_title('rvs')
    axs[2].hist(unif_rvs, normed=True, bins=50)     
    axs[2].set_ylim(a, b)
    axs[2].set_xlim(a, b)
    plt.show()
rivn(0, 2)
rivn(0, 5)
rivn(0, 10)
rivn(-2, 2)
rivn(-5, 5)
```

```python
# Місце для Вашого коду
from scipy.stats import norm
def n_r (mu, sigma):
    loc = mu
    scale = sigma
    fig, axs = plt.subplots(1,3, figsize=(20,5))
    print(fig)
    
    x2 = np.arange(-5, 5, 0.01)
    y2 = norm.pdf(x2, loc=loc, scale=scale)

    axs[0].plot(x2, y2, 'r')
    axs[0].set_title('pdf')
    axs[0].set_ylim(-5, 5)
    axs[0].set_xlim(-5, 5)
    y21 = norm.cdf(x2, loc=loc, scale=scale)

    axs[1].plot(x2, y21, 'g-')
    axs[1].set_title('cdf')
    axs[1].set_ylim(-5, 5)
    axs[1].set_xlim(-5, 5)
    norm_rvs = norm.rvs(loc=loc, scale=scale, size=100000)
    
    
    axs[2].plot(x2, y2, 'r-.')
    axs[2].set_title('rvs')
    axs[2].hist(norm_rvs, normed=True, bins=50)     
    axs[2].set_ylim(-5, 5)
    axs[2].set_xlim (-5, 5)
    plt.show()

n_r(0, 0.5)
n_r(0, 2)
n_r(0, 4)
n_r(0, 1)
n_r(-2, 1)
n_r(2, 1)
```

```python
# Місце для Вашого коду
from scipy.stats import norm
from scipy.stats import binom
n=100
list_p=[0.05, 0.1, 0.15, 0.2, 0.35, 0.5, 0.75]
for p in list_p:
    mu = n*p #мат.сподівання
    sigma = (n*p*(1-p))**0.5 #D=sigma**2
    loc = mu
    scale = sigma
    x2 = np.arange(0, 100, 0.1)
    x2g = np.arange(0, 100, 1)
    y2 = norm.pdf(x2, loc=loc, scale=scale)
    y2_ =norm.cdf(x2, loc=loc, scale=scale)
    fig, axs = plt.subplots(1,2, figsize=(20,5))
    yg = binom.pmf(x2g, n, p)
    yg_ = binom.cdf(x2g, n, p)
    
    axs[0].plot(x2, y2, 'r--', x2g, yg, 'b')
    axs[0].set_title('pdf')
    axs[1].plot(x2, y2_, 'r--', x2g, yg_, 'b')
    axs[1].set_title('cdf')
```

## 5 Апроксимація числа $\pi$

Розглянемо квадрат зі стороною $2a$. Нехай коло радіуса $a$ вписане в цей квадрат. Якщо вибрати точку рівномірно з цього квадрата, то ймовірність, що ця точка також належить і вписаному колу, дорівнює відношенню площ кола та квадрата, тобто $\pi a^2/(2a)^2 = \pi/4$.

Реалізуйте функцію емпіричної апроксимації числа $\pi$ за $n$ симуляцій (вибору $n$ випадкових точок з квадрата). Зобразіть графічно отримані елементи: квадрат, коло, випадкові точки. Виділіть точки, що належать колу, окремим кольором від інших точок.

Нехай $a = 1$. Виведіть результат функції для:
1. $n = 10$
2. $n = 100$
3. $n = 500$
4. $n = 1000$
5. $n = 2000$

```python
# Місце для Вашого коду
def pi(n):
    xk=np.arange(0, 2, 0.01)
    yk=np.arange(0, 2, 0.01)
    v_koli=0
    plt.plot(xk, xk*0, 'r')
    plt.plot(xk, xk*0+2, 'r')
    plt.plot(xk, 1+(-(xk-1)**2+1)**0.5, 'b')
    plt.plot(xk, 1-(-(xk-1)**2+1)**0.5, 'b')
    plt.plot(yk*0, yk, 'r')
    plt.plot(yk*0+2, yk, 'r')
    #plt.show()
    for i in range (n):
        x=2*np.random.sample(size=1)
        y=2*np.random.sample(size=1)
        if ((x-1)**2+(y-1)**2<=1):
            v_koli+=1
            plt.plot(x, y, 'b*')
        else:
            plt.plot(x, y, 'r*')
    plt.show()
    return(4*v_koli/n)

print(pi(10))        
print(pi(100))
print(pi(500)) 
print(pi(1000))
print(pi(2000))
```
